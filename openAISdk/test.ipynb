```python
agent = Agent(name="tool_agent", instructions="generate jokes about different topics.")
product_tool_joke = agent.as_tool(tool_name="any-role", tool_description="Use this tool to get a joke about a specific topic. 'any-role'.")
tool=[product_tool_joke]
tool
```

```python
agent = Agent(name="joke teller", instructions="your job is to tell a joke about developers, product and managers.", tools=tool, )
result = await Runner.run(agent, "Tell me 3 jokes about product, managers and developers.")
print(result.final_output)
```

```python

result = await asyncio.gather(
    Runner.run(agent, "Write a joke about developers."),
    Runner.run(agent, "Write a joke about product."),
    Runner.run(agent, "Write a joke about managers.")
)
output=[r.final_output for r in result]
joks= "\n---\n".join(output)
print(joks)
```

```python
agent = Agent(name="judge", instructions="you need to select the best joke among the provided options." )
```

```python
best_jok=Runner.run(agent, f"Here are some jokes:\n{joks}\nSelect the best one and explain why.")    
```
